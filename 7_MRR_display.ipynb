### Static

```python
writes, reads, qas = [], [], []
writes2, reads2, qas2 = [], [], []
ids = []
# filelist = os.listdir(dir_) 
for f in filelist:
    static_session = int(uuid_static_map[f.replace('.json', '')].replace('.json', '').split('_')[-1])
    try:
        data = json.load(open(os.path.join(dir_, f)))
    except:
        print(os.path.join(dir_, f))
        continue
    write = []
    for d1 in data:
        if d1['source_session_id'] > static_session:
            continue
        population = min(d1['valid_through_session'], static_session) - d1['source_session_id'] + 1
        target_session_scores1 = [ses for ses in d1['session_scores'] if ses['session_id'] <= static_session]
        multiply1 = population / len(target_session_scores1)
        
        write += [ses['write'] for ses in target_session_scores1] * int(round(multiply1))
        
    writes.append(np.mean(write))
    ids.append(f)
df = pd.DataFrame({'ids':ids, 'writes': writes})

display(df.describe().iloc[:3, :])
display(df)
```

### Dynamic

```python
writes, reads, qas = [], [], []
writes2, reads2, qas2 = [], [], []
ids = []
# filelist = os.listdir(dir_) 
for f in filelist:
    try:
        data = json.load(open(os.path.join(dir_, f)))
    except:
        print(os.path.join(dir_, f))
        continue
    write = []
    for d1 in data:
        population = d1['valid_through_session'] - d1['source_session_id'] + 1
        target_session_scores1 = [ses for ses in d1['session_scores']]
        multiply1 = population / len(target_session_scores1)
        write += [ses['write'] for ses in target_session_scores1] * int(round(multiply1))

    writes.append(np.mean(write))
    ids.append(f)
    
df = pd.DataFrame({'ids':ids, 'writes': writes})

display(df.describe().iloc[:3, :])
display(df)
```
